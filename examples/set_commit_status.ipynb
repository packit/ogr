# Setting commit statuses

```python
from ogr import ForgejoService

TOKEN = "<paste-your-token-here>"
instance_url = "https://codeberg.org"
namespace = "owner"
repo = "repo"

service = ForgejoService(instance_url=instance_url, token=TOKEN)
project = service.get_project(namespace=namespace, repo=repo)
```

Get the hash of the head commit of a specific PR:

```python
pr_id = 1000
pr = project.get_pr(pr_id=pr_id)
head_commit = pr.head_commit
```

Set a status flag on this commit:

```python

from ogr.abstract import CommitStatus

target_url = f"{instance_url}/{namespace}/{repo}"

project.set_commit_status(
    commit=head_commit,
    state=CommitStatus.success,
    target_url=target_url,
    description="Arbitrary status",
    context="System context",
)
```
